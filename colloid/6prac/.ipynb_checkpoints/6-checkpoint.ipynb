```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import UnivariateSpline
```

```python
# Constants
g = 9.8 # m / s^2
eta = 0.001 # Pa * s
H = 0.19 # m
l_density = 998.2 # kg / m^3
s_density =2700 # kg / m^3
```

```python
data = pd.read_csv('6.csv')
data.columns = ['time (s)', 'm (g)']
data.index = list(range(1, len(data)+1))
data.index.name = '№'
```

```python
data['P (N)'] = data['m (g)'] * g / 1000
```

### Экспериментальные данные (SiO2)

```python
data.head()
```

# Седиментационная кривая

```python
figure = plt.figure(figsize=(10, 7))

ax = sns.lineplot(data=data, x='time (s)', y='P (N)')

ax.set_title('Sedimentation curve')

t_min = 90
t_max = 1000

plt.axvline(x=t_min, color = 'red')
plt.axvline(x=t_max, color = 'green')

plt.text(t_min-30, 0.0025, 't min', fontsize=12, rotation=90, ha='center')

plt.text(t_max-30, 0.0025, 't max', fontsize=12, rotation=90, ha='center')

plt.show()
```

Накопление осадка в процессе оседания описывается уравнением Сведберга–Свен–Одена:

$$P_i = q_i + \left(\frac{dP}{dt}\right)_i t_i \tag{6.5}$$

где $P_i$ – общий вес осадка, соответствующий времени $t_i$, $q_i$ – вес частиц, полностью осевших к данному времени $t_i$, т.е. частиц, имеющих радиус больший, чем $r_i$; $\left(\frac{dP}{dt}\right)_i$ – вес частиц, радиус которых равен и меньше, чем $r_i$, и которые определяют скорость $\left(\frac{dP}{dt}\right)_i$ оседания в данный момент времени $t_i$.

Скорость оседания частиц радиуса $r_i$ равна $v_i = H/t_i$ и с учетом уравнения (6.4) $r_i$ рассчитывается как:

$$r_i = \sqrt{\frac{9\eta H}{2(\rho - \rho_0)gt_i}} = K\sqrt{\frac{H}{t_i}} \tag{6.6}$$

где $\eta$ – вязкость дисперсионной среды (для воды $1 \cdot 10^{-3}$ Па·с).

4. Рассчитать фракционный состав суспензии по кривой накопления графическим методом, используя уравнения (6.5) и (6.6). С этой целью на кривой накопления выбрать 8–10 точек с наибольшей кривизной и определить значения $t_i$, $q_i$ и $P_i$, соответствующие этим точкам: $q_i$ находят как отрезок, отсекаемый на оси ординат касательной, проведенной к зависимости $P(t)$ в точке с координатами $(t_i, P_i)$.

```python
data_filtered = data[(data['time (s)'] >= t_min) & (data['time (s)'] <= t_max)].copy()

# Cплайн-интерполяция (кубическая) для гладкой кривой.
spline = UnivariateSpline(data_filtered['time (s)'], data_filtered['P (N)'], s=0, k=3)

spline_derivative = spline.derivative(n=1)

selected_times = np.array([90, 100, 200, 300, 400, 500, 600, 700, 800, 900])

selected_P = spline(selected_times)

selected_dP = spline_derivative(selected_times)

# 6. Для каждой выбранной точки находим производную и q_i
tangent_data = []

for t_i, P_i, dP_dt in zip(selected_times, selected_P, selected_dP):

    # Уравнение касательной: P = P_i + dP_dt * (t - t_i)
    # При t = 0: q_i = P_i - dP_dt * t_i
    q_i = P_i - dP_dt * t_i
    
    tangent_data.append({
        't (s)': t_i,
        'P (N)': P_i,
        'dP_dt': dP_dt,
        'q (N)': q_i
    })

```

```python
figure = plt.figure(figsize=(14, 8))
ax = plt.subplot(111)

ax.plot(data['time (s)'], data['P (N)'], 'b-', linewidth=2, label='Experimental curve')

ax.scatter(selected_times, selected_P, color='red', s=100, zorder=5, label='Selected times')

# Tangents
for i, td in enumerate(tangent_data):
    t_range = np.linspace(0, data['time (s)'].max(), 100)
    tangent_line = td['P (N)'] + td['dP_dt'] * (t_range - td['t (s)'])
    ax.plot(t_range, tangent_line, '--', alpha=0.5, linewidth=1)
    
    ax.plot(0, td['q (N)'], 'o', markersize=8, color='green')

ax.axvline(x=t_min, color='red', linestyle=':', alpha=1, label='t min')
ax.axvline(x=t_max, color='green', linestyle=':', alpha=1, label='t max')
ax.set_xlim(0, data['time (s)'].max())
ax.set_ylim(bottom=min(0, min([td['q (N)'] for td in tangent_data])), top=0.006)
ax.set_xlabel('Time (s)', fontsize=12)
ax.set_ylabel('P (N)', fontsize=12)
ax.legend(loc='best')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

table = pd.DataFrame(tangent_data)
table.index = range(1, len(table) + 1)
table.index.name = '№'
```

### Экспериментальные данные, полученные из анализа кривой накопления.

```python
table
```

#### Вычисляем радиус частиц

```python
K = np.sqrt(9 * eta * H / (2 * (s_density - l_density) * g))

table['r (mkm)'] = 1e6 * K / np.sqrt(table['t (s)'])
table
```

### Интегральная кривая

Вычисляем содержание частиц Q в % от Pb max.

```python
table['Q (%)'] = 100 * table['q (N)'] / table['P (N)'].max()
table
```

```python
sns.lineplot(data=table, x='r (mkm)', y='Q (%)')
plt.show()
```

### Дифференциальная кривая

```python
diff_data = []

for i in range(len(table) - 1):
    r1 = table.iloc[i]['r (mkm)']
    r2 = table.iloc[i+1]['r (mkm)']
    Q1 = table.iloc[i]['Q (%)']
    Q2 = table.iloc[i+1]['Q (%)']
    
    r_mean = (r1 + r2) / 2  # средний радиус
    delta_r = abs(r1 - r2)  # интервал радиусов
    delta_Q = abs(Q1 - Q2)  # изменение содержания
    dQ_dr = delta_Q / delta_r if delta_r != 0 else 0  # дифференциальная функция
    
    diff_data.append({
        'r (mkm)': r_mean,
        'Δr (mkm)': delta_r,
        'ΔQ (%)': delta_Q,
        'ΔQ/Δr (%/mkm)': dQ_dr
    })

table3 = pd.DataFrame(diff_data)
table3.index = range(1, len(table3) + 1)
table3.index.name = '№'
display(table3)
```

```python
# Дифференциальная кривая
ax2 = plt.subplot(1, 2, 2)
ax2.plot(table3['r (mkm)'], table3['ΔQ/Δr (%/mkm)'], 'o-', 
         linewidth=2, markersize=8, color='red')
ax2.fill_between(table3['r (mkm)'], table3['ΔQ/Δr (%/mkm)'], 
                  alpha=0.3, color='red')
ax2.set_xlabel('Радиус частиц r, mkm', fontsize=12)
ax2.set_ylabel('ΔQ/Δr, %/mkm', fontsize=12)
ax2.set_title('Дифференциальная кривая распределения\nчастиц по размерам', fontsize=13)
ax2.grid(True, alpha=0.3)
ax2.set_ylim(bottom=0)

plt.tight_layout()
plt.show()
```
