```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### Работа № 7. Золь серы.

![](sol.jpg)

```python
data = pd.read_csv('7.csv')
data['lg(lambda)'] = np.log10(data['lambda (nm)'])
data['lg(A)'] = np.log10(data['A'])
data
```

```python
feature = data['lg(lambda)'].values.reshape(-1, 1)
y =  data['lg(A)'].values

model = LinearRegression()
model.fit(feature, y)
```

```python
x = np.linspace(2.6, 2.8, 10)

ax = sns.lineplot(data=data, x='lg(lambda)', y='lg(A)', marker='o', label='experimental data')
ax.plot(x, model.coef_*x + model.intercept_, label='regression')
print(f'x: {model.coef_}')
plt.legend()
plt.show()
```
